```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Importing the dataset
dataset_train = pd.read_csv('DAIC_WOZ_train.csv')
dataset_test = pd.read_csv('DAIC_WOZ_test.csv')

# Check for NaN values in the entire DataFrame
nan_check_train = dataset_train.isnull().values.any()
nan_check_test = dataset_test.isnull().values.any()

if nan_check_train:
    print("There are NaN values in the training dataset.")
    nan_counts = dataset_train.isnull().sum()
    print("NaN counts in each column of training dataset:\n", nan_counts)
else:
    print("There are no NaN values in the training dataset.")

if nan_check_test:
    print("There are NaN values in the test dataset.")
    nan_counts = dataset_test.isnull().sum()
    print("NaN counts in each column of test dataset:\n", nan_counts)
else:
    print("There are no NaN values in the test dataset.")

# Creating Dependent and Independent Variables
X_train = dataset_train.iloc[:, 1:88].values
y_train = dataset_train.iloc[:, -1].values

X_test = dataset_test.iloc[:, 1:88].values
y_test = dataset_test.iloc[:, -1].values

# Apply RandomOverSampler to balance the entire dataset
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X_train, y_train)

# Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_resampled)
X_test = sc.transform(X_test)

# Support Vector Machine (SVM)
svm_classifier = SVC(kernel='rbf', random_state=42)
svm_classifier.fit(X_train, y_resampled)
svm_y_pred = svm_classifier.predict(X_test)

# Random Forest
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_resampled)
rf_y_pred = rf_classifier.predict(X_test)

# Decision Trees
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_resampled)
dt_y_pred = dt_classifier.predict(X_test)

# Logistic Regression
lr_classifier = LogisticRegression(random_state=42)
lr_classifier.fit(X_train, y_resampled)
lr_y_pred = lr_classifier.predict(X_test)

# Evaluate each classifier
classifiers = {'SVM': svm_y_pred, 'Random Forest': rf_y_pred, 'Decision Trees': dt_y_pred, 'Logistic Regression': lr_y_pred}

for clf_name, clf_y_pred in classifiers.items():
    # Confusion Matrix
    cm = confusion_matrix(y_test, clf_y_pred)
    print(f"{clf_name} Confusion Matrix:\n", cm)

    # Accuracy Score
    accuracy = accuracy_score(y_test, clf_y_pred)
    print(f"{clf_name} Accuracy:", accuracy)

    # Classification Report
    report = classification_report(y_test, clf_y_pred)
    print(f"{clf_name} Classification Report:\n", report)
    print("="*50)
```
